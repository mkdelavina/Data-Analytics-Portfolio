```python
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
```

```python
#Example hypothesis: maybe the people on the upper decks were more likely to survive!
#(For reference, the decks are ordered in alphabetical order, so deck A was at the top, and G at the bottom of the ship).
sns.barplot(data=titanic, x = 'deck', y = "survived")
```

As we can see, it seems like there is not much of a correlation between the deck and the survival rate! (In case you are curious, the column "deck" is actually missing a lot of entries. You can confirm this by running::
 titanic.isna().sum()

 So, just in the case of using the deck, the database is not really reliable.)

**Hypothesis 1:** Men had better chances of survival because they are physically stronger.  <br>
**Conclusion:** Contrary to the hypothesis, the data shows that a big percentage of males, particularly those older than 15, did not survive. In my opinion it because women and children were prioritized during evacuation and rescue.


```python
#Work on your hypothesis 1 here
with_age = titanic.query('age >= 0 ')
sns.scatterplot(data=with_age, y="survived", x="age",hue="adult_male")
```

```python
sns.scatterplot(data=with_age, y="survived", x="age",hue="sex")
```

**Hypothesis 2:** Passenger class can predict passenger survival.  <br>
**Conclusion:** The hypotheis appears to be true. The data shows that on average, first class passengers had the highest survival rate, followed by second class, and third class has the lowest survival rate.

```python
#Work on your hypothesis 2 here
sns.lineplot(data=titanic, y="survived", x="pclass")

```

```python
sns.barplot(data=titanic, y="survived", x="pclass")
```

**Hypothesis 3:** Passengers who were alone had lower survival rate.  <br>
**Conclusion:** Hypothesis is true in general.Those who were alone had lower survival rate than those who were not. However, data shows that those with more than 1 sibling/spouse has lower survival rate on average. It is less conclusive when it comes to having parents and children.
